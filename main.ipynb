```python
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/statlog/german/german.data"
column_names = [
    "Status", "Duration", "CreditHistory", "Purpose", "CreditAmount", "Savings",
    "EmploymentSince", "InstallmentRate", "PersonalStatusSex", "OtherDebtors",
    "ResidenceSince", "Property", "Age", "OtherInstallmentPlans", "Housing",
    "NumberExistingCredits", "Job", "NumberPeopleLiable", "Telephone", "ForeignWorker", "Target"
]

# Veriyi oku
df = pd.read_csv(url, sep=' ', header=None, names=column_names)

# Hedef değişkeni 1: good, 2: bad olarak ayarlayalım
df['Target'] = df['Target'].map({1: 1, 2: 0})

df.to_csv("german_credit_data.csv", index=False)
df.head()
```

```python
df = df.rename(columns={
    'Status': 'Banka Hesap Durumu',
    'Duration': 'Kredi Süresi (Ay)',
    'CreditHistory': 'Kredi Geçmişi',
    'Purpose': 'Kredi Amacı',
    'CreditAmount': 'Kredi Miktarı',
    'Savings': 'Tasarruf Hesabı',
    'EmploymentSince': 'İş Süresi',
    'InstallmentRate': 'Taksit Oranı',
    'PersonalStatusSex': 'Cinsiyet/Medeni Durum',
    'OtherDebtors': 'Diğer Borçlular',
    'ResidenceSince': 'İkamet Süresi',
    'Property': 'Mülk Durumu',
    'Age': 'Yaş',
    'OtherInstallmentPlans': 'Diğer Taksit Planları',
    'Housing': 'Konut Durumu',
    'NumberExistingCredits': 'Mevcut Kredi Sayısı',
    'Job': 'Meslek Durumu',
    'NumberPeopleLiable': 'Bakmakla Yükümlü Kişi Sayısı',
    'Telephone': 'Telefon Durumu',
    'ForeignWorker': 'Yabancı İşçi',
    'Target': 'Kredi Onayı'  
})
df.head()
```

Sayısal olmayan kategorik sütunlar için label encoding uygulayalım. 

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
df['Banka Hesap Durumu'] = le.fit_transform(df['Banka Hesap Durumu'])
```

```python
df['Kredi Geçmişi'] = le.fit_transform(df['Kredi Geçmişi'])
df['Kredi Amacı'] = le.fit_transform(df['Kredi Amacı'])
df['Tasarruf Hesabı'] = le.fit_transform(df['Tasarruf Hesabı'])
df['İş Süresi'] = le.fit_transform(df['İş Süresi'])
df['Cinsiyet/Medeni Durum'] = le.fit_transform(df['Cinsiyet/Medeni Durum'])
df['Diğer Borçlular']= le.fit_transform(df['Diğer Borçlular'])
df['Mülk Durumu'] = le.fit_transform(df['Mülk Durumu'])
df['Diğer Taksit Planları'] = le.fit_transform(df['Diğer Taksit Planları'])
df['Konut Durumu'] = le.fit_transform(df['Konut Durumu'])
df['Meslek Durumu'] = le.fit_transform(df['Meslek Durumu'])
df['Telefon Durumu'] = le.fit_transform(df['Telefon Durumu'])
df['Yabancı İşçi'] = le.fit_transform(df['Yabancı İşçi'])
```

Eğitim-Test Verisi

```python
from sklearn.model_selection import train_test_split

x = df.drop("Kredi Onayı", axis = 1 )
y = df["Kredi Onayı"]

# Kategorik sütunlar
categorical_cols = x.select_dtypes(include=['object']).columns

# One-hot encoding ile kategorik sütunları sayısallaştır
x_encoded = pd.get_dummies(x, columns=categorical_cols, drop_first=True)

# Train-test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_encoded, y, test_size=0.2, random_state=42)

```

ML Modeli

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "SVM": SVC()
}
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
for name, model in models.items():
  try:
      model.fit(x_train, y_train)
      y_prediction = model.predict(x_test)

      acc = accuracy_score(y_test, y_prediction)
      print(f"\n {name}")
      print(f"Accuracy: {acc:.4f}")
      print("Confusion Matrix:")
      print(confusion_matrix(y_test, y_prediction))
      print("Classification Report:")
      print(classification_report(y_test, y_prediction))
  except Exception as e :
    print(f"{name} Modeli Yanıt Vermiyor: {e}")  
```

```python
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
y_pred = rf.predict(x_test)

print("Random Forest Accuracy:", accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
knn = KNeighborsClassifier()
knn.fit(x_train_scaled, y_train)
y_pred = knn.predict(x_test_scaled)
print("KNN Accuracy:", accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
svm_model = SVC()
svm_model.fit(x_train_scaled, y_train)
y_prediction = svm_model.predict(x_test_scaled)
print("SVM Accuracy:", accuracy_score(y_test, y_prediction))
print(confusion_matrix(y_test, y_prediction))
print(classification_report(y_test, y_prediction))
```
